```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
uniform_data = np.random.uniform(
    low=0,
    high=10,
    size=1000
)
```

```python
normal_data = np.random.normal(
    loc=5,
    scale=2,
    size=1000
)
```

```python
print("Uniform Distribution")
print("Mean =", np.mean(uniform_data))
print("Standard Deviation =", np.std(uniform_data))
```

```python
print("Normal Distribution")
print("Mean =", np.mean(normal_data))
print("Standard Deviation =", np.std(normal_data))
```

```python
plt.figure(figsize=(8,5))
plt.hist(uniform_data, bins=20)

plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(normal_data, bins=20)

plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

Uniform Distribution:
Values are spread almost equally across the range.

Normal Distribution:
Most values are concentrated around the mean.

Expectation:
The expected value represents the long-run average outcome.
