```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 + 2*x + 1
```

```python
def numerical_derivative(x, h=0.0001):
    return (f(x + h) - f(x)) / h
```

```python
points = [0, 2, 4]

for p in points:
    print(f"Derivative at x={p}: {numerical_derivative(p)}")
```

```python
x = np.linspace(-5, 5, 200)
y = f(x)

plt.figure(figsize=(10,6))
plt.plot(x, y, label='f(x)=x²+2x+1')

for p in points:
    slope = numerical_derivative(p)
    y_point = f(p)

    tangent = slope*(x-p) + y_point

    plt.plot(x, tangent, '--', label=f'Tangent at x={p}')
    plt.scatter(p, y_point)

plt.title("Curve and Tangent Lines")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

Observation 1:
The slope increases as x increases.

Observation 2:
Positive slope means the function is increasing.

Observation 3:
Near-zero slope indicates we are close to an optimum point in ML optimization.
