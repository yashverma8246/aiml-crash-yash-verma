```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing
```

```python
housing = fetch_california_housing(as_frame=True)

df = housing.frame

print("Dataset Loaded Successfully")
df.head()
```

```python
df.corr()
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['MedHouseVal'],
    bins=30,
    kde=True
)

plt.title("Distribution of House Values")
plt.xlabel("House Value")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['MedInc'],
    bins=30,
    kde=True
)

plt.title("Distribution of Median Income")
plt.xlabel("Median Income")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='MedInc',
    y='MedHouseVal',
    data=df
)

plt.title("Median Income vs House Value")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x=df['HouseAge']
)

plt.title("Boxplot of House Age")

plt.show()
```

```python
sns.pairplot(
    df[['MedInc',
        'HouseAge',
        'AveRooms',
        'MedHouseVal']]
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df['HouseAge'],
    bins=20
)

plt.title("House Age Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df['AveRooms'],
    bins=20
)

plt.title("Average Rooms Distribution")

plt.show()
```

```python
corr = df.corr()['MedHouseVal'].sort_values(ascending=False)

print(corr)
```
