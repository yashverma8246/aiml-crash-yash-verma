```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def gradient_descent(theta, lr, iterations):

    results = []

    for i in range(iterations):

        loss = (theta - 3.5)**2 + 1
        gradient = 2*(theta - 3.5)

        results.append([i, theta, loss, gradient])

        theta = theta - lr*gradient

    return pd.DataFrame(
        results,
        columns=['Iteration','Theta','Loss','Gradient']
    )
```

```python
theta_start = np.random.randint(-10,10)
print("Starting Theta =", theta_start)
```

```python
low_lr = gradient_descent(theta_start, 0.01, 15)
low_lr
```

```python
medium_lr = gradient_descent(theta_start, 0.1, 15)
medium_lr
```

```python
high_lr = gradient_descent(theta_start, 0.8, 15)
high_lr
```

```python
plt.figure(figsize=(10,6))

plt.plot(low_lr['Iteration'],
         low_lr['Loss'],
         label='LR = 0.01')

plt.plot(medium_lr['Iteration'],
         medium_lr['Loss'],
         label='LR = 0.1')

plt.plot(high_lr['Iteration'],
         high_lr['Loss'],
         label='LR = 0.8')

plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Loss Curve Comparison")
plt.legend()
plt.grid()
plt.show()
```

LR = 0.01:
Very slow convergence.

LR = 0.1:
Fast and stable convergence.

LR = 0.8:
May oscillate and become unstable.
