```python
# This program searches contacts in a contact book.

contacts = [
    {"name": "Manish", "phone": "9876543210", "email": "manish@gmail.com"},
    {"name": "Aniket", "phone": "9876543211", "email": "aniket@gmail.com"},
    {"name": "Palak", "phone": "9876543212", "email": "palak@gmail.com"},
    {"name": "Riyanshi", "phone": "9876543213", "email": "riyanshi@gmail.com"},
    {"name": "Yash", "phone": "9876543214", "email": "yash@gmail.com"}
]

def find_contact(name):
    for contact in contacts:
        if contact["name"].lower() == name.lower():
            return contact

    return "Contact not found"

search = input("Enter contact name: ")

print(find_contact(search))
```
