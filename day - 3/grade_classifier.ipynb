```python
# This program classifies students based on scores.

students = [
    {"name": "Yash", "score": 98},
    {"name": "Rahul", "score": 75},
    {"name": "Suman", "score": 61},
    {"name": "Riya", "score": 48},
    {"name": "Sunita", "score": 35}
]

def classify(score):
    if score >= 90:
        return "A"
    elif score >= 75:
        return "B"
    elif score >= 60:
        return "C"
    elif score >= 40:
        return "D"
    else:
        return "F"

sorted_students = sorted(students, key=lambda x: x["score"], reverse=True)

for student in sorted_students:
    print(f"{student['name']} - Grade: {classify(student['score'])}")
```
